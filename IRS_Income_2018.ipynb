```python
pivot_income = pd.pivot_table(data=df_nyc_income_2018, index='Zip Code', columns = 'Size of Adj Gross Income',values = 'Number of Returns',aggfunc='first')
```

```python
pivot_income
```

```python
pivot_income['Total Number of Tax Returns'] = pivot_income.sum(axis=1)
pivot_income
```

```python
pivot_income['WAVG Adj Gross Income']=((pivot_income['$1 under $25,000']*np.mean([1,25000])+pivot_income['$100,000 under $200,000']*np.mean([100000,200000])+pivot_income['$200,000 or more']*250000+pivot_income['$25,000 under $50,000']*np.mean([25000,50000])+pivot_income['$50,000 under $75,000']*np.mean([50000,75000])+pivot_income['$75,000 under $100,000']*np.mean([75000,100000]))/pivot_income['Total Number of Tax Returns']
```

```python
pivot_income
```

Categorize NY low, medium, high income:
low: < $50k ; medium: $50k < x < $150k ; high: > $150k


```python
pivot_income['Income Level']=''
pivot_income
```

```python
pivot_income.loc[(pivot_income['WAVG Adj Gross Income'] < 50000),'Income Level'] = 'Low'
```

```python
pivot_income.loc[(pivot_income['WAVG Adj Gross Income'] >= 50000) & (pivot_income['WAVG Adj Gross Income'] < 150000),'Income Level'] = 'Medium'
pivot_income.loc[(pivot_income['WAVG Adj Gross Income'] >= 150000),'Income Level'] = 'High'
```

```python
pivot_income.to_csv('Income.csv')
```
